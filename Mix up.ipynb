```python
from PIL import Image
import os
from random import randint
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_folder='D://mixup_data/'
images=os.listdir(img_folder)
batch_data=[Image.open(img_folder+p).resize((224,224)) for p in images]
```

```python
len(batch_data)
```

```python
def normalize_img(X):
    X=X/np.max(X)
    mean=np.mean(X)
    std=np.std(X)
    return ((X-mean)/std)
```

```python
def denormalize_img(X):
    
    mean=np.mean(X)
    std=np.std(X)
    X=X*std+mean
    X=np.clip(X,0,1)
    return X
```

```python
lamda=0.75
batch_size=len(batch_data)-1
current_img=normalize_img(np.array(batch_data[0]))
random_index=randint(0,batch_size)
random_img=normalize_img(np.array(batch_data[random_index]))
```

```python
plt.imshow(denormalize_img(current_img))
```

```python
plt.imshow(denormalize_img(random_img))
```

```python
mixed_image=lamda*current_img +(1-lamda)*random_img
```

```python
plt.imshow(denormalize_img(mixed_image))
```
